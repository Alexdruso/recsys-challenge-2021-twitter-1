```python
train=pd.read_csv("trainFinal.csv", index_col=0)
test=pd.read_csv("testFinal.csv", index_col=0)
```

```python
test.columns==train.columns
```

```python
columns=["retweet","retweetCom","reply","like"]

trainTarget= train[targetColName]
testTarget= test[targetColName]

for col in columns:
    train.drop(col,axis=1,inplace=True)
    test.drop(col,axis=1,inplace=True)
    
```

##Get data in correct format

```python
trainXGB=train.values
trainTargetXGB=trainTarget.values
del train
del trainTarget
```

```python
testXGB=test.values
testTargetXGB=testTarget.values
del test
del testTarget
```

#Model Building

##XGB computation

```python
xgb_params = { 
    'max_depth':8, 
    'learning_rate':0.025, 
    'subsample':0.85,
    'colsample_bytree':0.35, 
    'eval_metric':'logloss',
    'objective':'binary:logistic',
    'seed': 1,
}
```

```python
xgb_params["learning_rate"]=lr[targetIndex]
```

```python
xgbc = xgb.XGBClassifier(learning_rate=0.05,max_depth=8,colsample_bytree=0.5,seed=1,base_score=0.9)
```

```python
trained=xgbc.fit(trainXGB, trainTargetXGB,eval_metric="logloss",verbose=True)
```

```python
trainPred=trained.predict(trainXGB)
```

```python
cm = confusion_matrix(trainTargetXGB,trainPred)
cm
```

```python
testPred=trained.predict(testXGB)
```

```python
cm = confusion_matrix(testTargetXGB,testPred)
cm
```
